```python
# Создаем фигуру с несколькими подграфиками
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Плотность распределения
df['Age'].plot(kind='density', ax=axes[0,0], title='Плотность распределения возраста')
axes[0,0].set_xlabel('Возраст')
axes[0,0].set_ylabel('Плотность')

# Гистограмма
df['Age'].plot(kind='hist', ax=axes[0,1], title='Распределение возраста')
axes[0,1].set_xlabel('Возраст')
axes[0,1].set_ylabel('Количество')

# Столбчатая диаграмма (топ-10 возрастов)
df['Age'].value_counts().head(10).plot(kind='bar', ax=axes[1,0], title='Топ-10 возрастов')
axes[1,0].set_xlabel('Возраст')
axes[1,0].set_ylabel('Количество')

# Диаграмма рассеяния (возраст vs стоимость билета)
df.plot(kind='scatter', x='Age', y='Fare', ax=axes[1,1], title='Возраст vs Стоимость билета')
axes[1,1].set_xlabel('Возраст')
axes[1,1].set_ylabel('Стоимость билета')

# Круговая диаграмма распределения по классам
df['Pclass'].value_counts().plot(kind='pie', ax=axes[2,0], title='Распределение по классам')
axes[2,0].set_ylabel('')

# Областная диаграмма
df['Age'].plot(kind='area', ax=axes[2,1], title='Областная диаграмма возраста')
axes[2,1].set_xlabel('PassengerId')
axes[2,1].set_ylabel('Возраст')

plt.tight_layout()
plt.show()
```

```python
# Разделение таблицы по полу
df_male = df[df['Sex'] == 'male']
df_female = df[df['Sex'] == 'female']

# Заполнение пропущенных значений медианой для каждого пола
df_male.loc[:, 'Age'] = df_male['Age'].fillna(df_male['Age'].median())
df_female.loc[:, 'Age'] = df_female['Age'].fillna(df_female['Age'].median())

# Построение гистограммы распределения по возрасту
plt.figure(figsize=(12, 6))
plt.hist(df['Age'], bins=30, alpha=0.5, label='Все пассажиры')
plt.hist(df_male['Age'], bins=30, alpha=0.5, label='Мужчины')
plt.hist(df_female['Age'], bins=30, alpha=0.5, label='Женщины')
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.title('Распределение пассажиров по возрасту')
plt.legend()
plt.show()
```

```python
# Вычисление статистик по возрасту для каждого пола
stats_male = {
    'min': df_male['Age'].min(),
    'max': df_male['Age'].max(),
    'mean': df_male['Age'].mean(),
    'median': df_male['Age'].median(),
    'mode': df_male['Age'].mode()[0]
}

stats_female = {
    'min': df_female['Age'].min(),
    'max': df_female['Age'].max(),
    'mean': df_female['Age'].mean(),
    'median': df_female['Age'].median(),
    'mode': df_female['Age'].mode()[0]
}

print("Статистика по возрасту для мужчин:")
for key, value in stats_male.items():
    print(f"{key}: {value:.2f}")

print("\nСтатистика по возрасту для женщин:")
for key, value in stats_female.items():
    print(f"{key}: {value:.2f}")
```

```python
# Визуализация возраста разными способами

# 1. Box plot
df.boxplot(column='Age', by='Sex')
plt.title('Распределение возраста по полу (Box plot)')
plt.show()

# 2. Violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='Sex', y='Age', data=df)
plt.title('Распределение возраста по полу (Violin plot)')
plt.show()

# 3. Scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(df_male.index, df_male['Age'], alpha=0.5, label='Мужчины')
plt.scatter(df_female.index, df_female['Age'], alpha=0.5, label='Женщины')
plt.xlabel('Индекс пассажира')
plt.ylabel('Возраст')
plt.title('Распределение возраста (Scatter plot)')
plt.legend()
plt.show()

# 4. KDE plot
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df_male['Age'], label='Мужчины')
sns.kdeplot(data=df_female['Age'], label='Женщины')
plt.xlabel('Возраст')
plt.ylabel('Плотность')
plt.title('Распределение возраста (KDE plot)')
plt.legend()
plt.show()
```
